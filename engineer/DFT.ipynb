```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
# sampling rate
sr=100
# sampling interval
ts=1.0/sr
t=np.arange(0,1,ts)
print(len(t))
print(t)
```

```python
freq=1
x=3*np.sin(2*np.pi*freq*t)

freq=4
x+=np.sin(2*np.pi*freq*t)

freq=7
x+=0.5*np.sin(2*np.pi*freq*t)

plt.figure(figsize=(8,6))
plt.plot(t,x,'r')
plt.ylabel("Amplitude")

plt.show()
```

```python
def DFT(x):
    """
    Function to calculate the discrete
    Fourier Transform of a 1D 
    real-valued signal x
    """
    N=len(x)
    n=np.arange(N)
    k=n.reshape((N,1))

    e=np.exp(-2j*np.pi*k*n/N)
    print(e.shape,x.shape)
    X=np.dot(e,x)
    print(X.shape)
    return X
```

```python
X=DFT(x)
```

```python
# calculate the prequency
N=len(X)
n=np.arange(N)
T=N/sr
print(T)
freq=n/T
print(freq)

plt.figure(figsize=(8,6))
plt.stem(freq,abs(X),'b',markerfmt=" ",basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```
